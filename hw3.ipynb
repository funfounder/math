3.1 Напишите код, который будет переводить полярные координаты в декартовы.

```python
def pol2cart(radius, phi_angle):
    x = radius * np.cos(phi_angle)
    y = radius * np.sin(phi_angle)
    return([x,y])
```

Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
import numpy as np

def v_len(x):
    c_sq = 0
    for i in x:
        c_sq += int(i)**2
    return np.sqrt(c_sq)
    
x = [10, 10, 10]
print(v_len(x))
```

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы


```python
#парабола
x = np.arange(0, 2 * np.pi, 0.001)
r = 2
y = np.sqrt(2 * r * x)
y2 = -np.sqrt(2 * r * x)
plt.plot(x, y, x, y2)
plt.show()
```

```python
#окружность
#y^2+x^2=r^2
r = 6
x = np.arange(-r, r, 0.001)
y = np.sqrt(r**2-x**2)
plt.plot(x, y, x, -y)
plt.gca().set_aspect('equal')
plt.show()
```

```python
#элипс (x^2/a^2)+(y^2/b^2)=1
a = 2
b = 3
x = np.arange(-6, 6, 0.001)
y = np.sqrt((1-(x**2/a**2))*b**2)
plt.plot(x, y, x, -y)
plt.show()
```

```python
#гипербола (x^2/a^2)-(y^2/b^2)=1
a = 2
b = 3
x = np.arange(-6, 6, 0.001)
y = np.sqrt((1+(x**2/a**2))*b**2)
plt.plot(x, y, x, -y)
plt.show()
```
